# 서울시 공공자전거 대여 정보 분석
- 대여정보 읽기
    - encoding = 'cp949'
- 대여정보 확인 (결측치 컬럼 등)
- 대여일시, 반납일시를 날짜타입으로 변경
- 일자별로 대여 건수 내림차순으로 정렬하여 출력
    - Series객체.dt.날짜property
- 요일별 대여 건수 확인

## 대여 정보 데이터 읽기

```python
gong = pd.read_csv('서울특별시 공공자전거 대여정보_201911_2.csv', encoding = 'cp949')
```

## 대여정보 확인

```python
gong.head()
```

```python
gong.tail()
```

```python
gong.info()
```

## 대여일시, 반납일시를 날짜타입으로 변경

```python
gong['대여일시'] = pd.to_datetime(gong['대여일시'])
gong['반납일시'] = pd.to_datetime(gong['반납일시'])
# pd.to_datetime(data['기준시작일자'].replace('개통','',regex=True)
gong.info()
```

## 요일별 대여 건수 확인

```python
gong.head()

gong['요일'] = gong['대여일시'].dt.day_name()
```

```python
gong.head()
```

```python
gong.tail()
```

```python
gong.groupby('요일')[['대여일시']].count()
```

## 일자별로 대여 건수 내림차순으로 정렬하여 출력

```python
gong = gong.drop(['일자'], axis = 1)
```

```python
gong['일자'] = gong['대여일시'].dt.day
```

```python
gong.groupby('일자')[['대여일시']].count()
```

# 대여소 정보와 대여정보 Join
- 대여소 정보와 대여정보 join
- join 데이터 정보 확인
- 각 구별 이용시간과  이용거리에 대한 평균, 최대, 최소값 출력

```python
data.head()
```

```python
gong.head()
```

```python
gong.columns
```

## 대여소 ID 기준으로 두 데이터 병합

```python
total = pd.merge(data, gong, left_on = '대여소ID', right_on= '대여 대여소번호')
```

```python
total.head(5)
```

```python
total.columns
```

## 구별 평균 이용시간, 평균 이용거리, 최대 이용시간, 최소 이용시간 확인 (구 이름 정렬)

```python
total.groupby('대여소_구')[['이용시간','이용거리']].agg(['mean','max','min'])
```
