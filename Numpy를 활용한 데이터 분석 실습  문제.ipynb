# [문제] 국가건강검진 혈압혈당 데이터 분석

* '국가건강검진_혈압혈당데이터.csv' 파일 내용을 읽기 <data = np.loadtxt("국가건강검진_혈압혈당데이터.csv", delimiter=",", dtype=np.float, skiprows=1)>
    - 성별: 1(남자), 2(여자)
    - 연령그룹: 1(20-24), 2(25-26),...25-74세까지 2살 간격으로 1개 그룹으로 지정, 27(74세 이상)
    - 고혈압 당뇨병 진료여부: 1(고혈압 및 당뇨진료내역 있음), 2(고혈압 진료내역 있음), 3(당뇨 진료내역 있음), 4(고혈압 및 당뇨 진료내역 없음)
    - 정상혈압: 수축기 혈압 120nnHg 미만, 확장기 혈압 80mmHg 미만
    - 공복혈당 수치: 126 mg/dL 이상 -> 당뇨병, 100mg/dL-> 정상
* 전체 레코드 수와 상위 5개 데이터 확인
* 데이터 모양 확인
* 여성과 남성의 데이터 수 확인
* 여성과 남성의 수축기혈압과 이완기혈압의 평균값 확인
* 정상 혈압 남자 및 여자 수 확인
* 당뇨병 질환(공복혈당 수치 126mg/dL 이상)을 가지고 있는 사람들의 평균 체질량지수

```python
import numpy as np
data = np.loadtxt("국가건강검진_혈압혈당데이터.csv", delimiter=",", dtype=float, skiprows=1)
print(f'전체 레코드 수: {len(data):,}개')
print('탑 5:',data[:5])
print('데이터 모양 획인:', data.shape)
print('여성과 남성의 데이터 수 확인')
print(f'남성 수:{np.sum(data[: ,0]==1):,}개')
print(f'여성 수:{np.sum(data[: ,0]==2):,}개')
```

```python
# 각 컬럼별 unique  값 확인
print('성별 값: ', np.unique(data[:,0]))
# 성별의 컬럼은 1과 2의 데이터가 있다.

print('고열합,당뇨병 진료 여부 값', np.unique(data[:,5]) )
```

```python
# 수축기 혈압 : SBP
# 이완기 혈압 : DBP
print('남성의 수축기혈압 평균')
print(f'{np.mean(data[data[: ,0]==1], axis = 0)[2]:.2f}')
print(f'{np.mean(data[data[: ,0]==1][:,2]):.2f}')
print('남성의 이완기혈압 평균')
print(np.mean(data[data[: ,0]==1], axis = 0)[3])
print(f'{np.mean(data[data[: ,0]==1][:,3]):.2f}')

print('여성의 수축기혈압 평균')
print(f'{np.mean(data[data[: ,0]==2], axis = 0)[2]:.2f}')
print(f'{np.mean(data[data[: ,0]==2][:,2]):.2f}')
print('여성의 이완기혈압 평균')
print(np.mean(data[data[: ,0]==2], axis = 0)[3])
print(f'{np.mean(data[data[: ,0]==2][:,3]):.2f}')
#np.where(np.max(a)==a)
# 상혈압: 수축기 혈압 120nnHg 미만, 확장기 혈압 80mmHg 미만
man_s = data[data[: ,0]==1][:,2]
man_e = data[data[: ,0]==1][:,3]
print('정상 혈압 남자 수', np.sum(np.logical_and(man_s<120, man_e<80)))

woman_s = data[data[: ,0]==2][:,2]
woman_e = data[data[: ,0]==2][:,3]
print('정상 혈압 여자 수', np.sum(np.logical_and(woman_s<120, woman_e<80)))

#공복혈당 수치: 126 mg/dL 이상 -> 당뇨병, 100mg/dL-> 정상
print('당뇨병 질환(공복혈당 수치 126mg/dL 이상)을 가지고 있는 사람들의 평균 체질량지수:')
print(f'{np.mean(data[data[:,4]>=126][:,-1]):.2f}')
print(f'{np.mean(data[data[:,4]>=126], axis =0)[-1]:.2f}')
```

# [문제] 시에틀 강수량 데이터 분석
- 'Seattle2014.csv' 파일 내용 읽기
- 강수량('PRCP')이 많은 날 상위 10일 확인
- 1월 강수량만 출력
- 1월에 내린 강수량의 합, 평균강수량 확인
- 1년동안 강수량이 40미만으로 내린 날이 몇일인가?

```python
import numpy as np
data = np.loadtxt('Seattle2014.csv', delimiter=',', dtype = 'object', skiprows=1 )
#모든 데이터를 문자열로 읽어오기 위해서 dtype을 object로 지정
PRCP_Date =np.array(data[:,[2,3]], dtype = int)

print ("강수량('PRCP')이 많은 날 상위 10일 확인")
PRCP = PRCP_Date[:,1]
idx = np.argsort(PRCP)[::-1]
print(PRCP_Date[idx][:11])
#np.sort(data)[::-1] 


print('1월에 내린 강수량의 합, 평균강수량 확인')
print('방법 1')
Date = PRCP_Date [:,0]
print('합:',np.sum((PRCP_Date[Date<=20140131])[:,1]))
print('평균:',np.mean((PRCP_Date[np.logical_and(Date>=20140101,Date<=20140131)])[:,1]))

print('방법 2')
temp = np.array([[date[4:6],prcp]for date, prcp in data[:,[2,3]]])
jan_prcp = temp[temp[:,0]=='01'][:,1].astype(int)
print(jan_prcp)
print('합:',np.sum(jan_prcp))
print('평균:',np.mean(jan_prcp))

print('1년동안 강수량이 40미만으로 내린 날이 몇일인가')
PRC_ind = PRCP<40
print(np.sum(PRC_ind))
print(len(PRCP_Date[PRC_ind]))
#bolean 인덱싱 True and False
#필터링 된 데이터
```

# [문제] 영화 평점 분석
- 'ratings.dat' 파일 내용 읽기 (데이터 형식: user_id::item_id::rating::timestamp)
- 상위 10개 데이터 출력하기
- 데이터 정보 확인(크기, 차원)
- 전체 데이터 평점 구하기
- 사용자 아이디 1인 데이터 출력하기
- 사용자 아이디 1인 사람의 평점 평균 구하기
- 각 사용자별 평점 구하기 (사용자 ID 수집: ID컬럼에 대한 np.unique() 함수 적용)
- 각 사용자별 평균 평점이 4점 이상인 사용자 및 인원수 구하기

```python
import numpy as np
data = np.loadtxt('ratings.dat', delimiter='::', dtype = int)
print('상위 10개 데이터 출력하기')
ratings = data[:,2]
ratings_idx = np.argsort(ratings)[::-1]
print(ratings)
print(data[ratings_idx][:11])

```

```python
print('전체 데이터 평점 구하기')
print(f'{np.mean(ratings):.2f}')

print( '사용자 아이디 1인 데이터 출력하기')
user_id = data[:,0]
print(data[user_id==1])
```

```python
print('사용자 아이디 1인 사람의 평점 평균 구하기')
print(f'{np.mean(data[user_id==1][:,2]):.2f}')

```

```python
print('각 사용자별 평점 평균 구하기 (사용자 ID 수집: ID컬럼에 대한 np.unique() 함수 적용)')
for i in np.unique(user_id):
    print(f'id = {i}, 평점 평균 ={ np.mean(data[user_id == i ][:,2]):.2f}')

id_avg_rating = np.array([[id, np.mean(data[user_id == id][:,2])]for id in np.unique(user_id)])
#print(id_avg_rating)
```

```python
print('각 사용자별 평균 평점이 4점 이상인 사용자 및 인원수 구하기')

moreThan4 = []
for i in np.unique(user_id):
    mean = np.mean(data[user_id == i ][:,2])
    if mean >= 4:
        moreThan4.append(i)
print(f'각 사용자별 평균 평점이 4점 이상인 사용자: {moreThan4}')
print(f' 인원수 : {len(moreThan4)} 명')


print('각 사용자별 평균 평점이 4점 이상인 사용자:',np.sum(id_avg_rating[:,1]>=4))
```
