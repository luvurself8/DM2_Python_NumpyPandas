# 시애틀 강수량 데이터 분석
- 데이터 로딩
- 필요하지 않은 컬럼 삭제 (STATION, STATION_NAME)
- DATE를 날짜 형식으로 수정 : 기존 DATE 컬럼의 타입 int -> str로 변경 후 날짜 형식으로 변경
- 날짜 컬럼을 인덱스로 변경
- 강수량('PRCP')이 많은 날 상위 10일 출력
- 1월 강수량만 출력
- 1년동안 강수량이 40미만으로 내린 날이 몇일인가?
- 1월에 내린 강수량의 합과 평균 강수량
- 2월에 내린 강수량의 합과 평균 강수량

```python
import pandas as pd
import numpy as np

data = pd.read_csv('Seattle2014 (1).csv')
```

```python
data.head()
```

```python
#컬럼별 type 확인
data.info()
```

필요하지 않은 컬럼 삭제 (STATION, STATION_NAME)

```python
data.drop(['STATION', 'STATION_NAME'], axis = 1, inplace= True)
data
```

필요하지 않은 컬럼 삭제 (STATION, STATION_NAME)

```python
data['DATE'] = pd.to_datetime(data['DATE'].astype(str))
data['DATE']
```

날짜 컬럼을 인덱스로 변경

```python
df = pd.DataFrame(np.arange(12).reshape(6,2), index = [3,5,2,4,2,1])
df
```

```python
df = df.reset_index(drop=True) # 현재 인덱스를 삭제해라
df
```

```python
data.set_index('DATE', inplace = True)
data
```

강수량('PRCP')이 많은 날 상위 10일 출력

```python
data.sort_values('PRCP', ascending = False).head(10)[['PRCP']]
```

1월 강수량만 출력

```python
jan_prcp = data['PRCP'].iloc[:31]
jan_prcp
```

```python
month_list = list(data.index.month)
data['month'] = month_list
data

data.groupby('month').get_group(1)[['PRCP']]
```

```python
df = pd.DataFrame({'날짜': ['2022-01-01', '2022-02-01'],
                   '매출': [1000,2000]})
df['날짜'] = pd.to_datetime(df['날짜'])
df
```

```python
df['날짜'].month == 1
#시리즈 객채는 month 적용 못함
```

```python
df[df['날짜'].dt.month == 1]
#시리즈 객채는 month 적용하려면 dt 써야함
```

```python
data['PRCP'][data.index.month==1]
```

1년동안 강수량이 40미만으로 내린 날이 몇일인가

```python
sum(data['PRCP']<40)
```

```python
data[data['PRCP']<40].count() # 각 컬럼에 레코드 수
data[data['PRCP']<40].count()['PRCP']
```

```python
len(data[data['PRCP']<40])
```

1월에 내린 강수량의 합과 평균 강수량

```python
data['PRCP'][data.index.month ==1].agg(['sum','mean'])
```

```python
data.groupby('month')['PRCP'].agg([np.sum,np.mean])
```

```python
data.groupby('month').get_group(1)['PRCP'].agg([np.sum,np.mean])
```

2월에 내린 강수량의 합과 평균 강수량

```python
data.groupby('month').get_group(2)['PRCP'].agg([np.sum,np.mean])
```
